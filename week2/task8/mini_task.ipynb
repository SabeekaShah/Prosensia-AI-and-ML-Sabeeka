```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\New Moon\Documents\Prosensia_AI_ML_Internship\Prosensia-AI-and-ML-Sabeeka\week2\task6\features_v1.csv"
)

print(df.head())
print(df.info())
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['Month'] = df['order_date'].dt.month

df['DayOfWeek'] = df['order_date'].dt.day_name()

df['Is_Weekend'] = df['DayOfWeek'].isin(['Saturday', 'Sunday'])
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=['category', 'city', 'DayOfWeek'],
    drop_first=True
)
```

```python
df_encoded['Is_Weekend'] = df_encoded['Is_Weekend'].astype(int)
```

```python
df_encoded = df_encoded.drop(columns=['order_date'])
```
