```python
import pandas as pd
import numpy as np

np.random.seed(42)

df = pd.DataFrame({
    "Employee_ID": range(1, 101),
    "Age": np.random.randint(20, 60, 100),
    "Salary": np.random.randint(30000, 120000, 100),
    "Department": np.random.choice(
        ["HR", "IT", "Finance", "Marketing"],
        100
    )
})

# Add missing values
df.loc[5, "Age"] = np.nan
df.loc[12, "Salary"] = np.nan
df.loc[20, "Department"] = np.nan

# Add outliers
df.loc[30, "Salary"] = 500000
df.loc[40, "Age"] = 150

df.to_csv("corrupted_data.csv", index=False)

df.head()
```
