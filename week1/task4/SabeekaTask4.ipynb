```python
import pandas as pd
import numpy as np
import holidays
```

```python
df = pd.DataFrame({
    "Date": pd.date_range(start="2024-01-01", end="2024-12-31", freq="D")
})

df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["day_of_week"] = df["Date"].dt.day_name()
```

```python
df["month"] = df["Date"].dt.month_name()
```

```python
df["day_num"] = df["Date"].dt.dayofweek
df["month_num"] = df["Date"].dt.month
```

```python
pk_holidays = holidays.Pakistan()

df["is_holiday"] = df["Date"].dt.date.isin(pk_holidays).astype(int)
```

```python
day_encoded = pd.get_dummies(df["day_of_week"], prefix="day")
```

```python
month_encoded = pd.get_dummies(df["month"], prefix="month")
```

```python
df = pd.concat([df, day_encoded, month_encoded], axis=1)
```
