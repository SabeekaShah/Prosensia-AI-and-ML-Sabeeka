```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim

# set seeds for reproducibility
SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
```

```python
df = pd.read_csv("../../Week3/Weekend_Project/loan_default_dataset.csv")

print(f"raw dataset shape: {df.shape}")
print(df.head())
```

```python
df_clean = df.copy()

# fix inconsistent casing
df_clean["home_ownership"] = df_clean["home_ownership"].str.upper().str.strip()
df_clean["purpose"] = df_clean["purpose"].str.lower().str.strip()

# fix invalid ages
df_clean.loc[(df_clean["age"] < 18) | (df_clean["age"] > 100), "age"] = np.nan

# fill missing numeric values
numeric_cols = [
    "age", "annual_income", "credit_score", "loan_amount",
    "loan_term_months", "interest_rate", "debt_to_income",
    "employment_length_years", "num_open_accounts", "num_late_payments_2yrs"
]
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())

# one-hot encode categoricals
df_encoded = pd.get_dummies(df_clean, columns=["home_ownership", "purpose"], dtype=int)

print(f"cleaned and encoded shape: {df_encoded.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

TARGET = "default"
X = df_encoded.drop(columns=[TARGET])
y = df_encoded[TARGET]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=y
)

# fit scaler only on training data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"X_train: {X_train_scaled.shape}")
print(f"X_test : {X_test_scaled.shape}")
```

```python
# convert numpy arrays into PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train_scaled.copy())
y_train_tensor = torch.LongTensor(y_train.values.copy())

X_test_tensor = torch.FloatTensor(X_test_scaled.copy())
y_test_tensor = torch.LongTensor(y_test.values.copy())

print(f"X_train_tensor shape: {X_train_tensor.shape}, dtype: {X_train_tensor.dtype}")
print(f"y_train_tensor shape: {y_train_tensor.shape}, dtype: {y_train_tensor.dtype}")
print(f"\nsample tensor row:\n{X_train_tensor[0]}")
```

```python
class LoanDefaultMLP(nn.Module):
    def __init__(self, input_size):
        super(LoanDefaultMLP, self).__init__()
        # input layer -> hidden layer
        self.layer1 = nn.Linear(input_size, 32)
        self.relu = nn.ReLU()
        # hidden layer -> hidden layer
        self.layer2 = nn.Linear(32, 16)
        # output layer -> 2 classes (default / no default)
        self.output = nn.Linear(16, 2)

    def forward(self, x):
        x = self.layer1(x)
        x = self.relu(x)
        x = self.layer2(x)
        x = self.relu(x)
        x = self.output(x)
        return x

input_size = X_train_tensor.shape[1]
model = LoanDefaultMLP(input_size)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

print("loss function: CrossEntropyLoss")
print("optimizer: Adam, learning rate=0.001")
```

```python
epochs = 20
loss_history = []

for epoch in range(epochs):
    # zero the gradients from previous step
    optimizer.zero_grad()

    # forward pass
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    # backpropagation
    loss.backward()

    # optimizer step — update weights
    optimizer.step()

    loss_history.append(loss.item())
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, epochs+1), loss_history, marker="o", color="steelblue")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss Over Epochs")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("training_loss_curve.png")
plt.show()
print("loss curve saved — confirms convergence")
```

```python
model.eval()
with torch.no_grad():
    test_outputs = model(X_test_tensor)
    _, predicted = torch.max(test_outputs, 1)
    accuracy = (predicted == y_test_tensor).float().mean()

print(f"Test Accuracy: {accuracy.item():.4f}")
```

```python
explanation = """
=== WHY RELU OVER SIGMOID ===

Sigmoid: f(x) = 1 / (1 + e^(-x))
  Output range: (0, 1)
  Problem: For large positive or negative x, the curve
  flattens almost completely. Its derivative becomes
  extremely close to zero in these regions.

  During backpropagation, gradients are multiplied layer
  by layer (chain rule). If each layer's gradient is near
  zero, multiplying many small numbers together makes the
  gradient shrink exponentially as it travels backward
  through the network. This is the VANISHING GRADIENT
  PROBLEM — early layers stop learning almost entirely.

ReLU: f(x) = max(0, x)
  Output range: [0, infinity)
  For any x > 0, the derivative is exactly 1 — constant,
  not shrinking. Gradients pass through unchanged for all
  active (positive) neurons, no matter how deep the network.

  This means ReLU allows gradients to flow backward through
  many layers without shrinking to zero, making it possible
  to train much deeper networks effectively.

  Tradeoff: ReLU can cause "dead neurons" when x <= 0
  (gradient becomes exactly 0 there), but in practice this
  is far less damaging than the vanishing gradient problem
  that Sigmoid causes throughout the entire network.
"""
print(explanation)
```

```python
comparison = """
=== NUMPY ARRAYS vs PYTORCH TENSORS ===

NumPy Array:
  Static, CPU-only multidimensional array.
  No built-in tracking of operations performed on it.
  Used purely for numerical computation and storage.

PyTorch Tensor:
  Looks identical to a NumPy array on the surface, but
  carries an additional computational graph via autograd.
  Every operation performed on a tensor (with
  requires_grad=True) is recorded automatically.

  When .backward() is called, PyTorch walks backward
  through this recorded graph and computes gradients
  for every parameter automatically — this is what makes
  training neural networks possible without manually
  deriving every partial derivative by hand.

  Tensors can also be moved to GPU memory with .to('cuda')
  for massively parallel computation, something static
  NumPy arrays cannot do natively.
"""
print(comparison)
```
